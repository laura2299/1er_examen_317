```python
def generate_series(n):
    series = []
    if n==1:
      series.append(2)
    elif n==0:
      series.append(0)
    elif n % 2 == 0:
        for i in range(1, n // 2 + 1):
            series.append(i * i + 1)
            series.append(i + i)

    else:
        for i in range(1, n// 2 +1):
            series.append((i) * (i) + 1)
            series.append(i + i)
        series.append((i+1) * (i+1) + 1)  # Añadir el último elemento si n es impar

    return series


n = 1
resulting_series = generate_series(n)
print(resulting_series)
```

```python
from multiprocessing import Pool

def generar_serie_parcial(start, end):
    series = []
    for i in range(start, end):
        if i % 2 == 0:
            series.append(i * i + 1)
            series.append(i + i)
        else:
            series.append(i * i + 1)
            series.append(i + i)
    return series

def generar_serie(n, num_procesos=4):
    series = []

    if n == 1:
        series.append(2)
    elif n == 0:
        series.append(0)
    else:
        with Pool(processes=num_procesos) as pool:
            ranges = [(i, i + n // num_procesos) for i in range(1, n, n // num_procesos)]
            series_parciales = pool.starmap(generar_serie_parcial, ranges)
            for parcial in series_parciales:
                series.extend(parcial)

        if n % 2 == 1:
            series.append(n * n + 1)

    return series

n = 10000
resultado_serie = generar_serie(n)
print(resultado_serie)
```
